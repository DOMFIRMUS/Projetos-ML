```python
# importando a funcao para ler o pdf
from PyPDF2 import PdfReader
```

```python
# Lendo o arquivo PDF
leitor = PdfReader(open('/home/whitehat000142/Documents/Repositorios/Projetos-ML/Dados/Introducao_Analise_Real.pdf','rb'))
```

```python
# idenficando o numero de paginas do pdf 
numeros_paginas= len(leitor.pages)
print(f'{numeros_paginas} páginas')
```

```python
# identificando as informações da construcao do pdf 
informacoes = leitor.metadata
informacoes
```

```python
# variavel para armazenar o texto extraido do pdf
texto_extraido = ''

# loop para varrer o pdf
for pagina in range(numeros_paginas):
    #acessando a pagina do pdf de acordo com o loop 
    acessando_pagina = leitor.pages[pagina]

    #extraindo as informacoes
    extraindo_texto = acessando_pagina.extract_text()

    texto_extraido = texto_extraido + extraindo_texto


print(texto_extraido)
    
```

```python
# Gerando uma base analitica com as informacoes do pdf 

# vamos contar quantas palavras há em nosso pdf 
total_palavras = len(texto_extraido.split(' '))

dicionario = [{'creator': informacoes.creator,
               'producer': informacoes.producer,
               'creation_date': informacoes.creation_date,
               'total_palavras': total_palavras
              }]


from pandas import DataFrame

base = DataFrame(dicionario)
base
```
