```python
# Lib para Modelagem de dados
import pandas as pd
import numpy as np
```

```python
# Dados estão nessa URL
URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
```

```python
# Parametros para os dados
Colunas = ['symboling', 'normalized', 'make', 'fuel-type', 'aspiration',
           'num-of-doors', 'body', 'drive', 'engine-location', 'wheel', 'length',
           'width', 'height', 'curb', 'engine-type', 'num-of-cylinders', 'engine-size',
           'fuel-system', 'bore', 'stroke', 'compression', 'horsepower',
           'peak', 'city', 'highway', 'price']
```

```python
# Ler os dados
Base_Dados = pd.read_csv(URL, names=Colunas)
```

```python
# Verificando
Base_Dados.head()
```

```python
# Selecionar apenas colunas flutuantes para nosso exemplos
Colunas_Flutuantes = Base_Dados.columns[ Base_Dados.dtypes == 'float64' ]
```

```python
# Verificando
Base_Dados[Colunas_Flutuantes].head()
```

```python
# Função do MixMax
from sklearn.preprocessing import MinMaxScaler
```

```python
# Instanciando o MinMax
Funcao_MinMax = MinMaxScaler()
```

```python
# Aprender os parâmetros estatísticos para cada um dos dados e transformando
Dados_MinxMax = Funcao_MinMax.fit_transform(Base_Dados[Colunas_Flutuantes].values )
```

```python
# Verificando
pd.DataFrame(Dados_MinxMax,columns=Colunas_Flutuantes ).head()
```

```python
# Função do StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
# Instanciando o StandardScaler
Funcao_Standard = StandardScaler()
```

```python
# Aprender os parâmetros estatísticos para cada um dos dados e transformando
Dados_Standard = Funcao_Standard.fit_transform( Base_Dados[Colunas_Flutuantes].values )
```

```python
# Verificando
pd.DataFrame(Dados_Standard,columns=Colunas_Flutuantes ).head()
```

```python
# Função do RobustScaler
from sklearn.preprocessing import RobustScaler
```

```python
# Instanciando o RobustScaler
Funcao_RobustScaler = RobustScaler()
```

```python
# Aprender os parâmetros estatísticos para cada um dos dados e transformando
Dados_RobustScaler = Funcao_RobustScaler.fit_transform( Base_Dados[Colunas_Flutuantes].values )
```

```python
# Verificando
pd.DataFrame(Dados_RobustScaler,columns=Colunas_Flutuantes ).head()
```
