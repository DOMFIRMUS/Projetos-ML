```python
# Lib para Modelagem de dados
import pandas as pd
import numpy as np
```

```python
# Lib para analises gráficas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Base de Dados do Iris
Base_Dados = sns.load_dataset('iris')
```

```python
# Verificando
Base_Dados.head()
```

```python
# Aplicando a redução de dimensionalidade
# Reduzindo de 4 variaveis para 2
from sklearn.decomposition import PCA
```

```python
# Redução para 2 Variaveis
Funcao_PCA = PCA( n_components=2 )
```

```python
# Transformação dos Dados
Dados_PCA = Funcao_PCA.fit_transform( Base_Dados.drop( columns='species') )
```

```python
# Criando a Tabela
Base_PCA = pd.DataFrame( Dados_PCA, columns=['PC1','PC2'] )
```

```python
# Verificando
Base_PCA.head()
```

```python
# Plotando o gráfico com os dados 2 eixos
# Tamanho
plt.figure( figsize=(6, 5) )
# Titulo
plt.title('Divisão do PCA nos dados')
# Gráfico
sns.scatterplot( x='PC1', y='PC2', data=Base_PCA )
```

```python
# Plotando o gráfico com os dados 2 eixos
Base_PCA['species'] = Base_Dados['species']
# Tamanho
plt.figure( figsize=(6, 5) )
# Titulo
plt.title('Analisando a Classe')
# Gráfico
sns.scatterplot( x='PC1', y='PC2', hue='species', data=Base_PCA )
```
