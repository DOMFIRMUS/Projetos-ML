```python
# Lib para Modelagem de dados
import pandas as pd
import numpy as np
```

```python
# Ajuste no Pandas
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)
```

```python
# Leitura da base
Base_Dados = pd.read_csv('/home/whitehat000142/Documents/Repositorios/Projetos-ML/Dados/Base_Amostra_Enem_Bahia.csv')
```

```python
# Filtra apenas alunos de escolas públicas ou privadas
Base_Dados = Base_Dados[ Base_Dados.TP_ESCOLA.isin([2, 3]) ]
```

```python
# Filtra alunos que não obtiveram 0 em Matemática
# Filtra alunos com mais de 14 anos
Base_Dados = Base_Dados.query('(NU_NOTA_MT != 0) & (NU_IDADE >= 14)')
```

```python
# Verificando
Base_Dados.head()
```

```python
# Criando um datafreme vazio
Base_Modelada = pd.DataFrame()
```

```python
# Sexo do estudadnte
Base_Modelada['dummy_sexo_feminino'] = Base_Dados.TP_SEXO.map({'F': 1, 'M': 0} )
```

```python
# Idade
Base_Modelada['idade'] = Base_Dados.NU_IDADE
```

```python
# Etnia
Base_Modelada['cor_raca'] = Base_Dados['TP_COR_RACA']
```

```python
# Tipo Escola
Base_Modelada['escola_publica'] = Base_Dados['TP_ESCOLA'].map( {3:0, 2: 1} )
```

```python
# Nota de Matematica
Base_Modelada['nota_matematica'] = Base_Dados['NU_NOTA_MT']
```

```python
# Escolaridade do Pai
Base_Modelada['escolaridade_pai'] = Base_Dados.Q001.map({'A': 0, 'B': 0, 'C': 0, 'D': 0, 'E': 1, 'F': 1, 'G': 1, 'H': 0})
```

```python
# Escolaridade da Mãe
Base_Modelada['escolaridade_mae'] = Base_Dados.Q002.map({'A': 0, 'B': 0, 'C': 0, 'D': 0, 'E': 1, 'F': 1, 'G': 1, 'H': 0})
```

```python
# Tamanho da familia
Base_Modelada['tamanho_familia'] = Base_Dados.Q005
```

```python
# Renda Familiar
values_to_map = { value:index for index, value in enumerate(np.sort(Base_Dados.Q006.unique()), start = 1)}
Base_Modelada['renda_familiar'] = Base_Dados.Q006.map(values_to_map)
```

```python
# Possui internet
Base_Modelada['internet'] = Base_Dados.Q025.map({'A': 0, 'B': 1})
```

```python
# Possui Computador
Base_Modelada['computador'] = Base_Dados.Q024.map({'A': 0, 'B': 1, 'C': 1, 'D': 1, 'E': 1})
```

```python
# Retirando nulos
Base_Modelada.dropna(inplace = True)
```

```python
# Separar os dados
Caracteristicas = Base_Modelada.drop( columns='nota_matematica' )
Previsor = Base_Modelada['nota_matematica']
```

```python
# Divisão dos de Treino e Teste
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(Caracteristicas, Previsor,test_size=0.25, random_state=10)
```

```python
# Importar o Modelo
from sklearn.linear_model import LinearRegression
```

```python
# Definir a Regressão
Funcao_Regressao = LinearRegression()
Funcao_Regressao.fit( x_treino, y_treino )
```

```python
# Previsões
Previsoes = Funcao_Regressao.predict( x_teste )
```

```python
# Score do Modelo
Score_Modelo = Funcao_Regressao.score( x_treino, y_treino )
print( f'Score do modelo R2 {Score_Modelo}' )
```

```python
# Importando as metricas
from sklearn import metrics
# Calcular o RMSE
print('Erro Quadrático Médio (RMSE):',( metrics.mean_squared_error( y_teste, Previsoes, squared=False) ) )
```
