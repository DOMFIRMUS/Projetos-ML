```python
# Importar as libs necessarias

#libs para web scraping 

# Request é uma lib usada para request https
import requests

# soup é a lib usada para scraping 
from bs4 import BeautifulSoup


# Lib para modelagem de dados 
import pandas as pd

```

```python
# Carregando a pagina

# salvar o link da pagina
site = 'https://pt.wikipedia.org/wiki/Lista_de_municípios_de_Santa_Catarina'

# Fazendo o carregamento da página atraves do request
pagina = requests.get(site)
```

```python
# coletando as infos do request e passar para soup os dados
coleta = BeautifulSoup(pagina.text,'html.parser')
coleta
```

```python
# Coleta o titulo da pagina
print(coleta.title)

# Coletando apenas o texto 
print(coleta.title.string)

# Extrair o nome da tag
print(coleta.title.name)
```

```python
# Buscando uma classe pelo nome dela 
busca = coleta.find(class_='p-search--show-thumbnail')
print(busca)
```

```python
# Retornar todas as tags da pagina com um parametro 
print(coleta.p)
print(coleta.div)
#print(coleta.a)
```

```python
# Identificar o nome da classe de uma tag 
print(coleta.a['class'])
print(coleta.span['class'])
```

```python
# Identificar todas as tags com um parametro 
coleta.find_all('a')
```

```python
# procurar um id especifico da pagina 
coleta.find(id='p-search')
```

```python
# percorrer em um loop para coletar os links da pagina
for link in coleta.find_all('a'):
    print(link.get('href'))
```

```python
# coleta apenas o que for texto da pagina
print(coleta.get_text())
```

```python
# coletando todas as tbody e transformando em uma lista 
tabelas = list(coleta.find_all('tbody'))
```

```python
# listas para auxiliar na construcao da base 
cidades = []
estados = []
id = []

# variavel para ser usada no 'for'
loop = 0 

# loop nas tags 'a'
for text in tabelas[1].find_all('a'):

    # extraindo apenas o texto da tag
    cidade = text.string

    # caso o valor venha vazio, será ignorado
    if cidade == None:
        pass
    # se o valor for uma cidade, entra nosso 'else'
    else:
        # salvando os dados nas listas
        cidades.append(cidade)
        estados.append('Santa Catarina')
        id.append(loop)
        
        # somando o loop para virar o id na base de dados 
        loop += 1

# criando um dicionario para estruturar os dados

dicionario = {'id':id, 'cidade':cidades,'estado':estados}
```

```python
# passando o dicionario como parametro no pandas
base_cidades = pd.DataFrame(dicionario,)
```
